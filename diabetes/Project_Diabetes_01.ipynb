# Projeto Diabetes

## Contexto <br>
A diabetes está entre as doenças crónicas mais prevalentes nos Estados Unidos, afetando milhões de americanos todos os anos e exercendo um encargo financeiro significativo sobre a economia. A diabetes é uma doença crónica grave em que os indivíduos perdem a capacidade de regular eficazmente os níveis de glicose no sangue e pode levar à redução da qualidade e esperança de vida. Depois que diferentes alimentos são decompostos em açúcares durante a digestão, os açúcares são então liberados na corrente sanguínea. Isso sinaliza ao pâncreas para liberar insulina. A insulina ajuda a permitir que as células do corpo usem os açúcares da corrente sanguínea para obter energia. O diabetes é geralmente caracterizado pelo fato de o corpo não produzir insulina suficiente ou ser incapaz de usar a insulina produzida com a eficácia necessária.

Complicações como doenças cardíacas, perda de visão, amputação de membros inferiores e doenças renais estão associadas a níveis cronicamente elevados de açúcar remanescentes na corrente sanguínea para pessoas com diabetes. Embora não haja cura para o diabetes, estratégias como perder peso, alimentação saudável, ser ativo e receber tratamentos médicos podem mitigar os danos desta doença em muitos pacientes. O diagnóstico precoce pode levar a mudanças no estilo de vida e a um tratamento mais eficaz, tornando os modelos preditivos para o risco de diabetes ferramentas importantes para as autoridades de saúde pública.

A escala deste problema também é importante reconhecer. Os Centros de Controle e Prevenção de Doenças indicaram que em 2018, 34,2 milhões de americanos tinham diabetes e 88 milhões tinham pré-diabetes. Além disso, o CDC estima que 1 em cada 5 diabéticos e cerca de 8 em cada 10 pré-diabéticos desconhecem o seu risco. Embora existam diferentes tipos de diabetes, a diabetes tipo II é a forma mais comum e a sua prevalência varia de acordo com a idade, escolaridade, rendimento, localização, raça e outros determinantes sociais da saúde. Grande parte do fardo da doença recai também sobre aqueles de estatuto socioeconómico mais baixo. A diabetes também representa um fardo enorme para a economia, com custos de diabetes diagnosticados de cerca de 327 mil milhões de dólares e custos totais com diabetes e pré-diabetes não diagnosticados a aproximarem-se de 400 mil milhões de dólares anualmente.

## O Dataset
O Sistema de Vigilância de Fatores de Risco Comportamentais (BRFSS) é uma pesquisa telefônica relacionada à saúde coletada anualmente pelo CDC. Todos os anos, o inquérito recolhe respostas de mais de 400.000 americanos sobre comportamentos de risco relacionados com a saúde, condições crónicas de saúde e utilização de serviços preventivos. É realizado todos os anos desde 1984. Para este projeto, foi utilizado um csv do conjunto de dados disponível no Kaggle para o ano de 2015. Este conjunto de dados original contém respostas de 441.455 indivíduos e possui 330 recursos. Esses recursos são perguntas feitas diretamente aos participantes ou variáveis ​​calculadas com base nas respostas individuais dos participantes.
